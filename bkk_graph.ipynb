```python
import osmnx as ox
```

```python
G = ox.graph_from_place(
    "Bangkok, Thailand",
    network_type="drive",
    custom_filter='["highway"~"motorway|trunk|primary|secondary"]'
)
ox.plot_graph(G, node_size=3, edge_linewidth=0.5)
print(G)
```

```python
next(iter(G.edges(data=True)))
```

```python
ox.stats.basic_stats(G)
```
