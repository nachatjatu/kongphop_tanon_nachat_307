```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_2021 = pd.read_csv('data/2021_incidents.csv')
df_2022 = pd.read_csv('data/2022_incidents.csv')
df_2023 = pd.read_csv('data/2023_incidents.csv')
df_main = pd.concat([df_2021, df_2022, df_2023], ignore_index=True)
```

```python
df_main["latitude"] = pd.to_numeric(df_main["latitude"])
df_main["longitude"] = pd.to_numeric(df_main["longitude"])
df_main["starttime"] = pd.to_datetime(df_main["starttime"], errors='coerce')
df_main = df_main[df_main['starttime'].notna()]
df_main["description_en"] = df_main["description_en"].astype('string')
df_main["road_en"] = df_main["road_en"].astype('string')
```

```python
df_main['year'] = df_main['starttime'].dt.year
df_main['month'] = df_main['starttime'].dt.month
df_main['day_of_week'] = df_main['starttime'].dt.dayofweek    # Monday=0, Sunday=7
df_main['hour'] = df_main['starttime'].dt.hour
df_main = df_main.drop_duplicates()
```
